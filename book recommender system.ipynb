POPULARITY BASED RECOMMENDATION SYSTEM

```python
ratings_with_books=books.merge(rating,on='ISBN')
```

```python
ratings_with_books
```

```python
ratings_with_books['Book-Title'].nunique()
```

```python
title_rate=ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
title_rate.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
title_rate.head()
```

```python
avg_rating=ratings_with_books.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
avg_rating
```

```python
popular=pd.merge(title_rate,avg_rating,how='inner',on='Book-Title')
```

```python
popular=popular[popular['num_rating']>250]
```

```python
popular_df=popular.nlargest(columns='avg_rating',n=50)
```

```python
popular_df.shape
```

```python
final_df=popular_df.merge(books,how='inner',on='Book-Title')
```

```python
final_df=final_df.drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_rating','avg_rating','Image-URL-M']]
```

COLLABORATIVE FILTERING BASED SYSTEM

users>200 and books_num_rating>50

```python
ratings_with_books['User-ID'].value_counts()
```

```python
user_num_rating=ratings_with_books.groupby('User-ID').count()['Book-Rating'].reset_index()
```

```python
user_num_rating
```

```python
user_num_rating=user_num_rating[user_num_rating['Book-Rating']>200]['User-ID']
```

```python
user_num_rating
```

```python
filtered_rating=pd.merge(ratings_with_books,user_num_rating,how='inner',on='User-ID')
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
y=y[y].index
```

```python
df=filtered_rating[filtered_rating['Book-Title'].isin(y)]
```

```python
df
```

```python
table=pd.pivot_table(data=df,index='Book-Title',columns='User-ID',values='Book-Rating',fill_value=0)
```

```python
table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(table)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index=np.where(table.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]    
    for i in similar_items:
        print(table.index[i[0]])
```

```python
recommend('Zoya')
```
